```python
from scipy import stats

def cramers_V(var1,var2) :
    crosstab = np.array(pd.crosstab(var1,var2, rownames=None, colnames=None)) 
    stat = stats.chi2_contingency(crosstab)[0] 
    obs = np.sum(crosstab) 
    mini = min(crosstab.shape)-1 
    return (stat/(obs*mini))

def get_cramers_df(data):
    rows = []
    for var1 in data:
        col = []
        for var2 in data:
            cramers = cramers_V(data[var1], data[var2]) 
            col.append(round(cramers,2))
        rows.append(col)
    cramers_results = np.array(rows)
    return pd.DataFrame(cramers_results, columns = data.columns, index= data.columns)
```

```python
cramers_df = get_cramers_df(data_encoded)

plt.figure(figsize=(16, 6))

heatmap = sns.heatmap(cramers_df, vmin=0, vmax=1, annot=True)

heatmap.set_title('Correlación entre variables categóricas - V de Cramer', fontdict={'fontsize':12}, pad=12)
plt.show()
```

## Análisis sobre variación de producción y superficies anuales

Estudiamos las variaciones de la produccion de año a año, y vimos que ayudan ampliamente a los modelos predictivos.

Para eso:

- Creamos una funcion para obtener el valor del año anterior de la variable deseada.
- Creamos otra funcion para obtener la diferencia entre el valor actual y el pasado de la variable deseada, una vez ejecutada la funcion anterior.
- Copiamos el dataframe original y visualizamos aquellos cambios más bruscos

```python
def crear_shift(df,col):
    df = df.copy()
    df[f'{col}_t-1'] = df.groupby(['id_finca','variedad','modo'])[col].shift()
    df[f'{col}_t-2'] = df.groupby(['id_finca','variedad','modo'])[f'{col}_t-1'].shift()
    return df
```

```python
def crear_diferencias(df,col):
    df = df.copy()
    df[f'dif_{col}'] = abs(df[f'{col}_t-1']-df[col])
    return df
```

```python
df_shift = df.copy()
df_shift = crear_shift(df_shift,'produccion')
df_shift = crear_shift(df_shift,'superficie')
df_shift = crear_diferencias(df_shift,'produccion')
df_shift = crear_diferencias(df_shift,'superficie')
df_shift = df_shift.dropna(subset=['dif_produccion','dif_superficie'])
```

## Feature engineering para exploración

Creamos funciones para generar variables nuevas para el dataset de entrenamiento, siempre atentos a no realizar data leakage

- df_for_fe: dataframe para feature engineering, sin el año 21
- above_mean_col: variable que indique si para X valor de una columna, la produccion fue mayor o menor a la media
- quantiles_col: parecida a la anterior, que indique si el id de la finca u otro id tiene una produccion en X quantil.

```python
def df_for_fe(data):
    transf = data[data.campaña != 21]
    return transf

def above_mean_col(data,col):
    df = df_for_fe(data)
    above_mean = pd.DataFrame(df.groupby(col)['produccion'].mean() > df.produccion.mean())
    above_mean = list(above_mean[above_mean['produccion']].index)
    data[f'{col}_above_mean'] = [1 if x in above_mean else 0 for x in data[col]]
    return data

def quantiles_col(data,col):
    df = df_for_fe(data)
    quantiles_df = pd.DataFrame(df.groupby(col)['produccion'].mean()).reset_index()
    Q1 = quantiles_df.produccion.quantile(0.25)
    Q2 = quantiles_df.produccion.quantile(0.5)
    Q3 = quantiles_df.produccion.quantile(0.75)
    
    quantiles_df[f'{col}_percentiles'] = [0 if x < Q1 else 1 if x < Q2 else 2 if x < Q3 else 3 for x in quantiles_df.produccion]

    data = data.merge(quantiles_df[[col,f'{col}_percentiles']], how='left', on = col) #Joineamos con el df transformado
    
    return data
```

Aplicamos above_mean_col para altitud y variedad, y quantiles_col a finca y zona

```python
for col in ['altitud','variedad']:
    df=above_mean_col(df,col)
    
for col in ['id_finca','id_zona']:
    df=quantiles_col(df,col)
```

Podemos ir un paso más allá y procesar directamente los calculos anuales, para evitar entregarle un número muy grande de columnas al modelo.

Para esto, generamos funciones que nos reemplacen las variables de clima originales por su suma o promedio.

```python
def eto_dict_generator(df):
    temp_strings = ['Gust','MSLP','Humidity','UVIndex','Visibility','WindSpeed',
                    'lluvia','nieve','TemperatureLocalAfternoon','TemperatureLocalOvernight']
    dic_temp = {}
    for string in temp_strings:
        columns = [col for col in df.columns if string in col]
        dic_temp[string] = columns
    return dic_temp


def eto_annual_sum(df,cols_list):
    dictionary = eto_dict_generator(df)
    dictionary = {key: dictionary[key] for key in cols_list}
    for name, data in dictionary.items():
            df[name] = df[data].sum(axis=1)
            df = df.drop(columns=data)
    return df


def eto_annual_mean(df,cols_list):
    dictionary = eto_dict_generator(df)
    dictionary = {key: dictionary[key] for key in cols_list}
    for name, data in dictionary.items():
            df[name] = df[data].mean(axis=1)
            df = df.drop(columns=data)
    return df
```

Asignamos a que columna calcularle el promedio y a cual la suma

Preparación del pipeline final

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

def get_preprocesing():

    model_num_cols = [
        'campaña',
        'superficie',
        'prod_shift_max',
        'prod_shift_change',
        'prod_shift_avg',
        'prod_he_var_zone_mean_hist_total',
        'prod_he_var_zone_std_hist_total',
        'prod_he_var_modo_zona_mean_shift1_total',
        "prod_he_var_modo_zona_change_total"
    ]

    return ColumnTransformer([

        ('Flag',OrdinalEncoder(handle_unknown='use_encoded_value',unknown_value=-1), ['sup_is_nan']),
        ('Zona_encoder',mt.TargetEncoder(), ['id_zona']),
        ('Zona_encoder_2',OrdinalEncoder(handle_unknown='use_encoded_value',unknown_value=-1), ['id_zona']),
        ('Variedad_encoder',OrdinalEncoder(handle_unknown='use_encoded_value',unknown_value=-1), ['variedad']),
        ('Modo_encoder',OrdinalEncoder(handle_unknown='use_encoded_value',unknown_value=-1), ['modo']),
        ('Numerics_encoder',StandardScaler(), model_num_cols),

        ('Temp_pipeline',Pipeline([('Temp_pipeline',StandardScaler()),('asdasd',PCA(n_components=3, random_state=seed))]), temp_cols),

        ('Under_over_scaler',StandardScaler(),[c for c in precipuo if "2Std" in c]),

        ('Precip_PCA',make_pipeline(StandardScaler(),PCA(n_components=2, random_state=seed)),
             [c for c in percip_cols if 'Sum' in c]
        )],
        remainder='drop'
    )
```

```python
prep = get_preprocesing()
model = RandomForestRegressor(
    random_state=seed,
    n_estimators=500,
    min_samples_leaf=4,
    n_jobs=-1,
    max_features=0.20,
    max_samples=0.8
)

m = make_pipeline(
    prep,
    model
)
```

no  explicables  pueden  llegar  a  no  utilizarse  nunca  aun  teniendo  una  muy  buena
